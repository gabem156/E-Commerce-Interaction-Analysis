## Cleaning Dataset

**Find The File Name**

```python
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
```

**Load the Dataset**

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/e-commerce-sales-data/E-commerce sales data 2024.csv')

df.head()
```

**Inspect The Dataset**

```python
df.info()
```

**Check Missing Values**

```python
df.isnull().sum()
```

**Remove the Empty Column**

```python
df = df.drop(columns=['Unnamed: 4'])
```

**Verify**

```python
df.columns
```

**Check for Duplicate Rows**

```python
df.duplicated().sum()
```

**Save Progress**

```python
df.to_csv('cleaning_checkpoint.csv', index=False)
```

**Remove Duplicates**

```python
df = df.drop_duplicates()
```

**Verify**

```python
df.duplicated().sum()
```

**Check Missing Values Again**

```python
df.isnull().sum()
```

**Remove Remaining Missing Values**

```python
df = df.dropna()
```

**Verify**

```python
df.isnull().sum()
```

**Check Dataset Shape**

```python
df.shape
```

**Standardize Text Values**

```python
df['Interaction type'].value_counts()
```

```python
df.shape
```

**Save the Final Clean Dataset**

```python
df.to_csv('user_engagement_cleaned.csv', index=False)
```

## Data Cleaning Summary

- Removed empty column (Unnamed: 4)
- Removed 294 duplicate records
- Removed rows containing missing values
- Converted Time stamp to datetime format
- Final dataset contains 2,871 records and 4 columns
